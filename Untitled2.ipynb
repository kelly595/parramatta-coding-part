```python
import pandas as pd
import numpy as np
import os
import shutil
from sqlalchemy import create_engine, text

folder_path = "/Users/happykelly/Desktop/2001ass/"
businesses_path = folder_path + "Businesses.csv"
income_path = folder_path + "Income.csv"
population_path = folder_path + "Population.csv"

print("Cleaning Businesses data...")
businesses_df = pd.read_csv(businesses_path)

businesses_df.columns = businesses_df.columns.str.lower()
businesses_df = businesses_df.dropna() 

numeric_cols = ['0_to_50k_businesses', '50k_to_200k_businesses', 
                '200k_to_2m_businesses', '2m_to_5m_businesses', 
                '5m_to_10m_businesses', '10m_or_more_businesses', 
                'total_businesses']

for col in numeric_cols:
    businesses_df[col] = pd.to_numeric(businesses_df[col], errors='coerce')
    businesses_df = businesses_df.dropna(subset=[col]) 
    businesses_df[col] = businesses_df[col].astype(int)

businesses_df['calculated_total'] = businesses_df[numeric_cols[:-1]].sum(axis=1)
total_mismatch = businesses_df[businesses_df['calculated_total'] != businesses_df['total_businesses']]

if not total_mismatch.empty:
    print(f"Found {len(total_mismatch)} rows with mismatched totals, fixing...")
    businesses_df['total_businesses'] = businesses_df['calculated_total']
businesses_df = businesses_df.drop('calculated_total', axis=1)

print(f"Businesses data cleaning completed, {len(businesses_df)} rows remaining")
print(businesses_df.head())

income_df = pd.read_csv(income_path)

income_df.columns = income_df.columns.str.lower()
income_df = income_df.dropna()  
income_df['sa2_code21'] = income_df['sa2_code21'].astype(str)

numeric_cols = ['earners', 'median_age', 'median_income', 'mean_income']
for col in numeric_cols:
    income_df[col] = pd.to_numeric(income_df[col], errors='coerce')
    
income_df = income_df.dropna(subset=numeric_cols)

for col in numeric_cols:
    income_df[col] = income_df[col].astype(int)

print(f"Income data cleaning completed, {len(income_df)} rows remaining")
print(income_df.head())


population_df = pd.read_csv(population_path)

population_df.columns = population_df.columns.str.lower().str.replace('-', '_')
population_df = population_df.dropna()  

population_df['sa2_code'] = population_df['sa2_code'].astype(str)

age_cols = [col for col in population_df.columns if 'people' in col]
for col in age_cols:
    population_df[col] = pd.to_numeric(population_df[col], errors='coerce')
population_df = population_df.dropna(subset=age_cols)


for col in age_cols:
    population_df[col] = population_df[col].astype(int)

calculated_total = population_df[[col for col in age_cols if col != 'total_people']].sum(axis=1)
total_mismatch = population_df[calculated_total != population_df['total_people']]

if not total_mismatch.empty:
    print(f"Found {len(total_mismatch)} rows with mismatched population totals, fixing...")
    population_df['total_people'] = calculated_total

print(f"Population data cleaning completed, {len(population_df)} rows remaining")
print(population_df.head())

backup_folder = folder_path + "original_backup/"

if not os.path.exists(backup_folder):
    os.makedirs(backup_folder)

shutil.copy2(businesses_path, backup_folder + "Businesses_original.csv")
shutil.copy2(income_path, backup_folder + "Income_original.csv")
shutil.copy2(population_path, backup_folder + "Population_original.csv")
print("Original files backed up to:", backup_folder)

businesses_df.to_csv(businesses_path, index=False)
income_df.to_csv(income_path, index=False)
population_df.to_csv(population_path, index=False)
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import json
import requests
import time  
from shapely.geometry import Point
from sqlalchemy import create_engine, text


def pgconnect(credential_filepath, db_schema="public"):
    with open(credential_filepath) as f:
        db_conn_dict = json.load(f)
        host       = db_conn_dict['host']
        db_user    = db_conn_dict['user']
        db_pw      = db_conn_dict['password']
        default_db = db_conn_dict['database']
        port       = db_conn_dict['port']
        try:
            db = create_engine(f'postgresql+psycopg2://{db_user}:{db_pw}@{host}:{port}/{default_db}', echo=False)
            conn = db.connect()
            print('Connection successful.')
        except Exception as e:
            print("Unable to connect to the database.")
            print(e)
            db, conn = None, None
        return db, conn

credentials_path = "Credentials.json"
db, conn = pgconnect(credentials_path)

try:
    parramatta_query = "SELECT * FROM parramatta"
    parramatta = gpd.read_postgis(parramatta_query, conn, geom_col='geometry')
    print(f"Loaded {len(parramatta)} Parramatta SA2 regions from database")
except Exception as e:
    print("Error loading Parramatta data:", e)
    
    shapefile_path = "/Users/happykelly/Desktop/SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp"
    print(f"Attempting to load from shapefile: {shapefile_path}")
    
    gdf = gpd.read_file(shapefile_path)
    gdf = gdf.set_crs('EPSG:7844', allow_override=True)
    gdf_sydney = gdf[gdf["GCC_NAME21"] == "Greater Sydney"]
    gdf_sydney = gdf_sydney.to_crs(epsg=4326)
    parramatta = gdf_sydney[gdf_sydney["SA4_NAME21"] == "Sydney - Parramatta"]
    parramatta = parramatta[~parramatta.geometry.isna()]
    print(f"Loaded {len(parramatta)} Parramatta SA2 regions from shapefile")

def get_arcgis_pois(minx, miny, maxx, maxy):
    url = "https://maps.six.nsw.gov.au/arcgis/rest/services/public/NSW_POI/MapServer/0/query"
    bbox_str = f"{minx},{miny},{maxx},{maxy}"
    params = {
        "f": "geojson",
        "geometryType": "esriGeometryEnvelope",
        "geometry": bbox_str,
        "inSR": "4326",
        "spatialRel": "esriSpatialRelIntersects",
        "outFields": "*",
        "returnGeometry": "true"
    }
    try:
        response = requests.get(url, params=params)
        data = response.json()
        return data.get("features", [])
    except Exception as e:
        print("API Error:", e)
        return []

print("Collecting Points of Interest data...")
poi_data = []
for idx, row in parramatta.iterrows():
    print(f"Processing SA2 region: {row['SA2_NAME21']} ({idx+1}/{len(parramatta)})")
    minx, miny, maxx, maxy = row.geometry.bounds
    features = get_arcgis_pois(minx, miny, maxx, maxy)
    print(f"  Found {len(features)} POIs")
    
    for feature in features:
        props = feature.get("properties", {})
        geom = feature.get("geometry", {})
        coords = geom.get("coordinates", [None, None])
        poi_data.append({
            "sa2_code": row["SA2_CODE21"],
            "sa2_name": row["SA2_NAME21"],
            "poi_name": props.get("NAME"),
            "poi_group": props.get("POI_GROUP"),
            "longitude": coords[0],
            "latitude": coords[1]
        })
    time.sleep(1)  
print(f"Processing {len(poi_data)} POIs...")
poi_df = pd.DataFrame(poi_data)

print("Check POI:")
for column in poi_df.columns:
   null_count = poi_df[column].isna().sum()
   print(f"  Column '{column}' has {null_count} missing values ({null_count/len(poi_df)*100:.1f}%)")
critical_columns = ['longitude', 'latitude', 'sa2_code']
poi_df = poi_df.dropna(subset=critical_columns)
print(f"Retained {len(poi_df)} POI records after dropping rows with missing critical columns")

poi_df = poi_df.fillna({
    'poi_name': 'Unknown POI',
    'poi_group': 'Other'
})

poi_df['geometry'] = poi_df.apply(lambda r: Point(r['longitude'], r['latitude']), axis=1)
poi_gdf = gpd.GeoDataFrame(poi_df, geometry='geometry', crs="EPSG:4326")

if 'poi_id' not in poi_gdf.columns:
    poi_gdf['poi_id'] = range(1, len(poi_gdf) + 1)
if 'poi_type' not in poi_gdf.columns:
    poi_gdf['poi_type'] = poi_gdf['poi_group']

print(f"Saving {len(poi_gdf)} POIs to database...")
poi_gdf.to_postgis("points_of_interest", db, if_exists="replace")
print("Points of Interest data saved to database successfully.")

poi_count_query = "SELECT COUNT(*) FROM points_of_interest"
poi_count = pd.read_sql_query(poi_count_query, conn)
print(f"Total POIs in database: {poi_count.iloc[0, 0]}")
```

```python
try:
    credentials_path = "Credentials.json"
    db, conn = pgconnect(credentials_path)
    print("Database connection established")
    
    try:
        parramatta_query = "SELECT * FROM parramatta"
        parramatta = gpd.read_postgis(parramatta_query, conn, geom_col='geometry')
        print(f"Loaded {len(parramatta)} Parramatta SA2 regions")
    except Exception as e:
        print(f"Error loading Parramatta data: {e}")
        try:
            shapefile_path = "/Users/happykelly/Desktop/SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp"
            gdf = gpd.read_file(shapefile_path)
            gdf = gdf.set_crs('EPSG:7844', allow_override=True)
            gdf_sydney = gdf[gdf["GCC_NAME21"] == "Greater Sydney"]
            gdf_sydney = gdf_sydney.to_crs(epsg=4326)
            parramatta = gdf_sydney[gdf_sydney["SA4_NAME21"] == "Sydney - Parramatta"]
            parramatta = parramatta[~parramatta.geometry.isna()]
            print(f"Loaded {len(parramatta)} Parramatta SA2 regions from shapefile")
        except Exception as e2:
            print(f"Error loading from shapefile: {e2}")
            raise Exception("Cannot proceed without Parramatta data")
    
    # 加载站点数据
    try:
        stops_query = "SELECT * FROM stops"
        stops_gdf = gpd.read_postgis(stops_query, conn, geom_col='geometry')
        print(f"Loaded {len(stops_gdf)} transit stops from database")
    except Exception as e:
        print(f"Error loading stops data from database: {e}")
        try:
            stops_file = "/Users/happykelly/Desktop/2001ass/Stops.txt"
            stops_df = pd.read_csv(stops_file)
            from shapely.geometry import Point
            stops_df['geometry'] = stops_df.apply(
                lambda row: Point(row['stop_lon'], row['stop_lat']), 
                axis=1
            )
            stops_gdf = gpd.GeoDataFrame(stops_df, geometry='geometry', crs="EPSG:4326")
            print(f"Loaded {len(stops_gdf)} transit stops from file")
        except Exception as e2:
            print(f"Error loading stops from file: {e2}")
            stops_gdf = gpd.GeoDataFrame()
    
    try:
        catchments_query = "SELECT * FROM school_catchments"
        catchments_gdf = gpd.read_postgis(catchments_query, conn, geom_col='geometry')
        print(f"Loaded {len(catchments_gdf)} school catchments from database")
    except Exception as e:
        print(f"Error loading catchments data from database: {e}")
        try:
            catchments_file = "/Users/happykelly/Desktop/2001ass/catchments/catchments_primary.shp"
            catchments_gdf = gpd.read_file(catchments_file)
            catchments_gdf = catchments_gdf.to_crs(epsg=4326)
            print(f"Loaded {len(catchments_gdf)} school catchments from file")
        except Exception as e2:
            print(f"Error loading catchments from file: {e2}")
            catchments_gdf = gpd.GeoDataFrame()
    
    print("\nPerforming spatial joins...")
    
    if not stops_gdf.empty:
        stops_with_sa2 = gpd.sjoin(stops_gdf, parramatta[['SA2_CODE21', 'geometry']], how='inner', predicate='intersects')
        stops_with_sa2 = stops_with_sa2.dropna(subset=['SA2_CODE21'])  
        stops_with_sa2['SA2_CODE21'] = stops_with_sa2['SA2_CODE21'].astype(str).str.zfill(9)
        stop_counts = stops_with_sa2.groupby('SA2_CODE21').size()
        print(f"Matched {len(stops_with_sa2)} stops to SA2 regions")
        print(f"Stops are distributed across {len(stop_counts)} SA2 regions")
    else:
        stop_counts = pd.Series(dtype=float)
        print("No stops data available for spatial join")
    
    if not catchments_gdf.empty:
        catchments_with_sa2 = gpd.sjoin(catchments_gdf, parramatta[['SA2_CODE21', 'geometry']], how='inner', predicate='intersects')
        catchments_with_sa2 = catchments_with_sa2.dropna(subset=['SA2_CODE21']) 
        catchments_with_sa2['SA2_CODE21'] = catchments_with_sa2['SA2_CODE21'].astype(str).str.zfill(9)
        school_counts = catchments_with_sa2.groupby('SA2_CODE21').size()
        print(f"Matched {len(catchments_with_sa2)} school catchments to SA2 regions")
        print(f"School catchments are distributed across {len(school_counts)} SA2 regions")
    else:
        school_counts = pd.Series(dtype=float)
        print("No catchments data available for spatial join")
    
    print("\nLoading business and POI data...")
    
    try:
        business_df = pd.read_sql("SELECT * FROM businesses", db)
        print(f"Loaded {len(business_df)} business records")
        
        parramatta_sa2_codes = parramatta['SA2_CODE21'].astype(str).str.zfill(9).tolist()
        
        business_df['sa2_code'] = business_df['sa2_code'].astype(str).str.zfill(9)
        
        business_df_filtered = business_df[business_df['sa2_code'].isin(parramatta_sa2_codes)]
        
        print(f"Filtered to {len(business_df_filtered)} business records in Parramatta region")
        print(f"Original business data covered {business_df['sa2_code'].nunique()} SA2 regions")
        print(f"Filtered business data covers {business_df_filtered['sa2_code'].nunique()} SA2 regions in Parramatta")
        
        business_counts = business_df_filtered.groupby('sa2_code')['total_businesses'].sum()
        business_counts.index = business_counts.index.astype(str)
        
        business_df = business_df_filtered
        
    except Exception as e:
        print(f"Error loading business data: {e}")
        business_counts = pd.Series(dtype=float)
        print("No business data available for counting")
    
    try:
        poi_df = pd.read_sql("SELECT * FROM points_of_interest", db)
        print(f"Loaded {len(poi_df)} points of interest")
        
        poi_df['sa2_code'] = poi_df['sa2_code'].astype(str).str.zfill(9)
        
        poi_df_filtered = poi_df[poi_df['sa2_code'].isin(parramatta_sa2_codes)]
        
        print(f"Filtered to {len(poi_df_filtered)} POIs in Parramatta region")
        print(f"POIs are distributed across {poi_df_filtered['sa2_code'].nunique()} SA2 regions in Parramatta")
        
        poi_counts = poi_df_filtered.groupby('sa2_code').size()
        poi_counts.index = poi_counts.index.astype(str)
        

        poi_df = poi_df_filtered
        
    except Exception as e:
        print(f"Error loading POI data: {e}")
        poi_counts = pd.Series(dtype=float)
        print("No POI data available for counting")
    
    if not stop_counts.empty:
        stop_counts.index = stop_counts.index.astype(str)
    else:
        print("No stop counts available")
    
    if not school_counts.empty:
        school_counts.index = school_counts.index.astype(str)
    else:
        print("No school counts available")
    
    print("\nCreating score dataframe...")
    parramatta_sa2_codes = parramatta['SA2_CODE21'].astype(str).str.zfill(9)
    score_df = pd.DataFrame(index=parramatta_sa2_codes)
    
    score_df['businesses'] = score_df.index.map(business_counts).fillna(0)
    score_df['stops'] = score_df.index.map(stop_counts).fillna(0)
    score_df['schools'] = score_df.index.map(school_counts).fillna(0)
    score_df['pois'] = score_df.index.map(poi_counts).fillna(0)
    
    try:
        population_df = pd.read_sql("SELECT * FROM population", db)
        population_df.columns = population_df.columns.str.strip().str.lower()
        population_df['sa2_code'] = population_df['sa2_code'].astype(str).str.zfill(9)
        
        young_cols = ['0_4_people', '5_9_people', '10_14_people', '15_19_people']
        for col in young_cols:
            if col not in population_df.columns:
                print(f"Warning: {col} column not found in population data")
                population_df[col] = 0
        
        population_df['young_people'] = population_df[young_cols].sum(axis=1)
        
        population_sub = population_df[population_df['sa2_code'].isin(parramatta_sa2_codes)].copy()
        
        score_df['young_people'] = score_df.index.map(
            population_sub.set_index('sa2_code')['young_people'].to_dict()
        ).fillna(0)
        
        score_df['total_people'] = score_df.index.map(
            population_sub.set_index('sa2_code')['total_people'].to_dict()
        ).fillna(0)
        
        print(f"Processed population data for {len(population_sub)} SA2 regions")
        print(f"Average young population: {score_df['young_people'].mean():.1f}")
        print(f"Average total population: {score_df['total_people'].mean():.1f}")
        
    except Exception as e:
        print(f"Error loading population data from database: {e}")
        print("Attempting to load from CSV file...")
        
        population_df = pd.read_csv("/Users/happykelly/Desktop/2001ass/Population.csv")
        population_df.columns = (
            population_df.columns
            .str.strip()
            .str.lower()
            .str.replace("-", "_")
            .str.replace(" ", "_")
            )
            
        population_df['sa2_code'] = population_df['sa2_code'].astype(str).str.zfill(9)
        population_df['young_people'] = population_df[[
            '0_4_people', '5_9_people', '10_14_people', '15_19_people'
        ]].sum(axis=1)
        
        population_sub = population_df[population_df['sa2_code'].isin(parramatta_sa2_codes)].copy()
            
        score_df['young_people'] = score_df.index.map(
            population_sub.set_index('sa2_code')['young_people'].to_dict()
        ).fillna(0)
            
        score_df['total_people'] = score_df.index.map(
            population_sub.set_index('sa2_code')['total_people'].to_dict()
        ).fillna(0)
            
        print(f"Processed population data from CSV for {len(population_sub)} SA2 regions")
        print(f"Average young population: {score_df['young_people'].mean():.1f}")
        print(f"Average total population: {score_df['total_people'].mean():.1f}")
            

    print("\nScore dataframe head:")
    print(score_df.head())
    
    print("\nMissing young people values:", score_df['young_people'].isna().sum())
    
    print("\nCalculating final scores...")
    
    score_df['businesses_per_1000'] = np.where(
        score_df['total_people'] > 0, 
        (score_df['businesses'] / score_df['total_people']) * 1000,
        0
    )
    
    score_df['stops_per_1000'] = np.where(
        score_df['total_people'] > 0,  
        (score_df['stops'] / score_df['total_people']) * 1000,
        0
    )
    
    score_df['schools_per_1000_young'] = np.where(
        score_df['young_people'] > 0,
        (score_df['schools'] / score_df['young_people']) * 1000,
        0
    )
    
    score_df['pois_per_1000'] = np.where(
        score_df['total_people'] > 0, 
        (score_df['pois'] / score_df['total_people']) * 1000,
        0
    )
    
    for col in ['businesses_per_1000', 'stops_per_1000', 'schools_per_1000_young', 'pois_per_1000']:
        mean_val = score_df[col].mean()
        std_val = score_df[col].std()
        if std_val > 0:  
            score_df[f'z_{col}'] = (score_df[col] - mean_val) / std_val
        else:
            score_df[f'z_{col}'] = 0
    
    
    z_total = (score_df['z_businesses_per_1000'] + 
               score_df['z_stops_per_1000'] + 
               score_df['z_schools_per_1000_young'] + 
               score_df['z_pois_per_1000'])
    
    score_df['total_score'] = 1.0 / (1.0 + np.exp(-z_total))
    
    
    print("\nTop 5 SA2 regions by score:")
    top_regions = score_df.sort_values('total_score', ascending=False).head(5)
    top_regions_with_names = top_regions.copy()
    sa2_names = parramatta.set_index(parramatta['SA2_CODE21'].astype(str).str.zfill(9))['SA2_NAME21']
    top_regions_with_names['sa2_name'] = top_regions_with_names.index.map(sa2_names)
    print(top_regions_with_names[['sa2_name', 'total_score']])
    
    
    try:
        score_df_for_db = score_df.reset_index()
        score_df_for_db.rename(columns={'index': 'sa2_code21'}, inplace=True)
        
        score_df_for_db.to_sql('sa2_scores', db, if_exists='replace', index=False)
        print("\nScores saved to database successfully.")
    except Exception as e:
        print(f"\nError saving scores to database: {e}")

except Exception as e:
    print(f"Critical error: {e}")
finally:
    if 'conn' in locals() and conn is not None:
        conn.close()
        print("\nDatabase connection closed.")
```
